## Reading data from dataset

```python
# For avoiding future warnings while training model
import pandas as pd
import numpy as np

import warnings
warnings.simplefilter('ignore', FutureWarning)
```

```python
# Reading data from CSV
data = pd.read_csv('concrete_data.csv')
data.head()
```

```python
# Specifying columns for the predictors and target values
data_columns = data.columns
predictors = data[data_columns[data_columns != 'Strength']]
target = data['Strength']
```

```python
target.head()
```

```python
predictors.head()
```

## Splitting the data

```python
# Using Train Test Split from SKLEARN LIBRARY
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    predictors, target, test_size=0.3, random_state=42)

# Training data = 70% and Testing Data = 30%
```

## Building a Neural Network

```python
# Importing the Keras Library for creating model
import keras
from keras.models import Sequential
from keras.layers import Dense
```

## Train and Test the Network

Here, we are creating a model which has 1 Hidden Layer of 10 nodes and Activation Function is ReLU. Optimizer is adam and Loss Function is mean_squared_error. We are storing the loss values in a list as well... there are 50 epochs

```python
# create model
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(predictors.shape[1],)))
model.add(Dense(1))

# compiling model with MSE as the Loss Function
model.compile(optimizer='adam', loss='mse')

# Train the model and store the history
history = model.fit(X_train, y_train, epochs=50)

# Storing the MSE values in the list
mse_list = history.history['loss']
```

```python
print(mse_list)
```

```python
# Mean value of the Losses throughout the Training
print(np.mean(mse_list))
```

```python
# Standard Deviation of the Losses throughout the Training
print(np.std(mse_list))
```
